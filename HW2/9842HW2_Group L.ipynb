### Problem 1###
\begin{eqnarray*}
L(x,y,\lambda)=(x-a)^2+(y-b)^2-\lambda_1 x+\lambda_2 (x-1)-\lambda_3 y +\lambda_4 (y-1)
\end{eqnarray*}
Then local minimizer should subject to

\begin{equation}  
\left\{  
             \begin{array}{ll}
             2(x^*-a)-\lambda_1^*+\lambda_2^*=0\\  
             2(y^*-a)-\lambda_3^*+\lambda_4^*=0\\
             \lambda_1^* x^*=0\\
             \lambda_2^*(x^*-1)=0\\
             \lambda_3^* y^*=0\\
             \lambda_4^*(y^*-1)=0\\
             \lambda_i^* \geq 0 & \text{for } i=1,2,3,4
             \end{array}  
\right.  
\end{equation}  

Solve the equations above, the solution is
\begin{equation}  
x^*=
\left\{  
             \begin{array}{ll}
             0 & a\leq 0\\
             a & a \in (0,1) \\
             1 & a \geq 1
             \end{array}  
\right.  
\end{equation} 
\begin{equation}  
y^*=
\left\{  
             \begin{array}{ll}
             0 & b\leq 0\\
             b & b \in (0,1) \\
             1 & b \geq 1
             \end{array}  
\right.  
\end{equation}  

### Problem 2###
We present two solutions below, with slightly different approach.

#### Solution 1
We replace constraints of the form
\begin{equation} |x_i| \leq b_i \end{equation}
with
\begin{equation}
\begin{split}
	x_i \leq b_i \\
	-x_i \leq b_i
\end{split}
\end{equation}

To deal with the absolute value in the objective function, we introduce a new variable U, s.t
\begin{equation}
\begin{split}
	x_2-10 \leq U \\
	10-x_2 \leq U
\end{split}
\end{equation}


Note that $U \geq |x_2-10|$ always holds. As $U$ has no other constrains, when we minimize $U$, $U=|x_2-10|$ always holds.

Thus, by change of variables and substitutions above, the nonlinear problem can be rewritten as LP problem
\begin{equation}
\min \, 2x_1+3U, \text{ subject to  }
\left\{
    \begin{array}{l}
    x_1 + 2 +x_2-10 \leq 5 \\
    -x_1 - 2 +x_2-10 \leq 5 \\
    x_1 + 2 +10-x_2  \leq 5 \\
    -x_1 - 2 +10-x_2  \leq 5 \\
    x_2-10 \leq U \\
	10-x_2 \leq U \\
    \end{array}
\right.
\end{equation}

#### Solution 2
Let $x_2-10=x^+-x^-,\, x^+,x^-\geq 0$, use the property of absolute value inequalities, the problem become
\begin{equation}  
\min \, 2x_1+3|x^+-x^-|, \text{ subject to}
\left\{  
             \begin{array}{ll}
             x_1+2+10+x^+-x^-\leq 5\\
             -x_1-2+10+x^+-x^-\leq 5\\
             x_1+2-10-x^++x^-\leq 5\\
             -x_1-2-10-x^++x^-\leq 5\\
             x^+,x^-\geq 0
             \end{array}  
\right.  
\end{equation}  

Notice that 
\begin{equation}  
2x_1+3(x^++x^-) \geq 2x_1+3|x^+-x^-|
\end{equation}  
now we consider 
\begin{equation}  
\min \,2x_1+3(x^++x^-) , \text{ subject to}
\left\{  
             \begin{array}{ll}
             x_1+2+10+x^+-x^-\leq 5\\
             -x_1-2+10+x^+-x^-\leq 5\\
             x_1+2-10-x^++x^-\leq 5\\
             -x_1-2-10-x^++x^-\leq 5\\
             x^+,x^-\geq 0
             \end{array}  
\right.  
\end{equation}  
Let $x_1^*,x^{+*},x^{-*}$ be the minimizer of the problem above, the next step is to prove that  $x_1^*,x^{+*},x^{-*}$ is the minimizer of the original problem, we prove by contradiction.
If there is $x_1',x_2'$ subject to the conditions and 
\begin{equation}  
2x_1'^*+3|x_2'-10| < 2x_1^*+3(x^{+*}+x^{-*})
\end{equation}  
Let 
\begin{eqnarray*}  
x_{1\star}=x_1'\\
x^{+}_\star=\{x_2'-10\}^+\\
x^{^-}_\star=\{-x_2'+10\}^+
\end{eqnarray*}  
Then 
\begin{equation}  
2x_{1\star}+3(x^{+}_\star+x^{^-}_\star) < 2x_1^*+3(x^{+*}+x^{-*})
\end{equation}  
Contradiction!

Recall equation (5) we know that
\begin{equation}  
2x_1'^*+3|x_2'-10| = 2x_1^*+3(x^{+*}+x^{-*})
\end{equation}  
So we can always use the minimizer of problem (6) to get the minimizer and minimum value of the original problem. Then problem has been transferred into a LP
\begin{equation}  
\min \,2x_1+3(x^++x^-) , \text{ subject to}
\left\{  
             \begin{array}{ll}
             x_1+2+10+x^+-x^-\leq 5\\
             -x_1-2+10+x^+-x^-\leq 5\\
             x_1+2-10-x^++x^-\leq 5\\
             -x_1-2-10-x^++x^-\leq 5\\
             x^+,x^-\geq 0
             \end{array}  
\right.  
\end{equation}  

### Problem 3###

There are $N$ currencies, index $i=1,\dots,N$.

We start with $B$ units of currency $1$ and we want to maximize the number
of units of currency $N$ through a sequence of trades. One unit of 
currency $i$ can be exchanged for $f_{ij}$ units of currency $j$.

Decision variables: let $x_{ij} =$ amount of currency $i$ to be converted 
into currency $j$.

The no-arbitrage condition 
$f_{i_1 i_2}\,f_{i_2 i_3}\dots f_{i_{k-1} i_k}\,f_{i_k i_1}\leq 1 $ means
that if we start with any currency, continually exchange one currency into another
currency, and eventually back into the original currency, we can never end up
with more units of the original currency, i.e. if we start with 1 unit, at
most we get back 1 unit, or could be less than 1, after exchanging into a series of
other currencies. It also implies there can be no trading 'loops' or trading 
sequences in which one can increase currency count beyond its initial allocation
in this economy. 
The trade from one currency to another currency happens at most for one time.

Clearly $\sum_{j\not= 1}x_{1j}\leq B$. We also have 
constraint $\sum_{j\not= 1}x_{1j}\leq u_1$, since we exchange out of 
currency 1 exactly once into currency $j$ ($j \not= 1,N$) and never trade it back to currency 1, 
where $u_i$ is the limit on the total amount of currency $i$ that can be
traded. If we exchange out of currency 1 into currency $N$ through 
intermediate currency $i$, considering currency  the tradings from other currency to currency $i$, the trading limit is
$(\sum_{j \not= 1}x_{ij} + \sum_{k \not= N}x_{ki}f_{ki}) \leq u_i$, for $i \not= 1,N$. The amount of buying currency $i$ should no less than selling it, so we have $\sum_{j\not= 1, i}x_{ij}\leq \sum_{k\not=i,N}f_{ki} x_{ki}$.

Objective function: maximize $\sum_{i \not= N} f_{iN} * x_{iN}$, essential
maximizing the total inflow into currency $N$ from all other currencies.
There is no term $x_{i1}$ nor $x_{Ni}$ for $i=1,\dots,N$ in our solution because they are definitely 0 and we have no need to include them. Any such positive value will definitely waste trading limit as well as reduce total value of our asset.

In summary, the LP formulation of this problem is

\begin{equation}  
\max \,\sum_{i \not= N} f_{iN} * x_{iN}, \text{ subject to}
\left\{  
             \begin{array}{ll}
             \sum_{j\not= 1}x_{1j}\leq B,\\
             \sum_{j\not= 1}x_{1j}\leq u_1,\\
             \sum_{j\not= 1, i}x_{ij}\leq \sum_{k\not=i,N}f_{ki} x_{ki},& \text{for } i\not= 1,N\\
             \sum_{j \not= 1}x_{ij} + \sum_{k \not= N}x_{ki}f_{ki} \leq u_i, &\text{for } i \not= 1,N\\
             x_{ij}\geq 0, & \text{for } i\not= j,i\not=N
             \end{array}  
\right.  
\end{equation}
